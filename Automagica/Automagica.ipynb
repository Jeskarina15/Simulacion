**AUTOMAGICA PPT APPLICATION**

En este ejemplo, crearemos una presentación de PowerPoint totalmente automatizada

```python
from automagica import *
##Iniciando PowerPoint
##Comience abriendo PowerPoint, en este caso le daremos a la instancia el nombre 'presentación'. 
automagicaPPT = PowerPoint()
##Crear nuestra primera diapositiva
automagicaPPT.add_slide()
##Ahora agregamos algo de contenido. 
automagicaPPT.add_text(text='Hola, esto es power point con automagica', index=1)
```

```python
##Guardar datos
##Ahora podemos guardar la presentación, 
##elegimos guardarla como .pdf y como .pptx. Finalmente no olvidemos salir de PowerPoint.

##Primero declaramos el path utilizando la función home_path integrada de Automagica para asegurarnos de saber dónde se guarda.
presentation_pptx = home_path('automagicaPPT.pptx')

presentation_pdf = home_path('automagicaPPT.pdf')

# Save pptx
automagicaPPT.save_as(presentation_pptx)
# Save pdf
automagicaPPT.export_to_pdf(presentation_pdf)
##Cerramos el documento
automagicaPPT.quit()
## Verificamos abriendo el contenido en formato PDF
open_file(presentation_pdf)
```
