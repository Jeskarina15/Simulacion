El centro de inters se mueve desde un modelo determinstico a uno probabilstico usando tcnicas estadsticas subjetivas para estimacin, prueba y prediccin.
```python
import simpy
import random
import matplotlib.pyplot as pp
import time
m=round(media)
DIAS_SIMULACION=224
trabajadores_alimentacion={}
trabajadores_retail={}
trabajadores_restaurante={}
trabajadores_viajes={}
trabajadores_hoteles={}
trabajadores_desempleados={}

class ReinsercionLaboral(object):
    def __init__(self, env, valor):
        self.env = env
        self.valor = simpy.Resource(env, valor)
    
    def reintegracion_persona(self, actividad):
        if actividad == "alimentacion":
            trabajadores_alimentacion[env.now]=trabajadores_alimentacion[env.now]+ 1 if env.now in trabajadores_alimentacion else  1
        elif actividad == "retail":
            trabajadores_retail[env.now]=trabajadores_retail[env.now]+ 1 if env.now in trabajadores_retail else  1
        elif actividad == "restaurante y ocio":
            trabajadores_restaurante[env.now]=trabajadores_restaurante[env.now]+ 1 if env.now in trabajadores_restaurante else  1
        elif actividad == "viajes":
            trabajadores_viajes[env.now]=trabajadores_viajes[env.now]+ 1 if env.now in trabajadores_viajes else  1
        elif actividad == "hoteles":
            trabajadores_hoteles[env.now]=trabajadores_hoteles[env.now]+ 1 if env.now in trabajadores_hoteles else  1
            
    def teletrabajo(inicio, trabajador, actividad):
        yield inicio.env.timeout(random.randint(m-2, m+5))
        ##Tiempo de teletrabajo de 5 a 14 dias, se puso de 2 a 3 porque se iba a demorar mucho con todos los datos
    ## pero se podra poner el rango de 5 a 14, pero la salida de datos se demoraria ms 
        tiempo= random.randint(2,3)
        time.sleep(tiempo)
        valoracion= random.randint(0,100)
        #####La persona tiene una probabilidad del 70% para reintegrarse nuevamente a sus labores presenciales#########
        if(valoracion>30):
            print("El trabajador recuperado ", trabajador, " del area ", actividad, "se integra a laborar de forma presencial")
            env.process(inicio.reintegracion_persona(actividad))
            return
        else: 
            print("El trabajador recuperado ", trabajador, " del area ", actividad, "vuelve al teletrabajo")
            env.process(inicio.teletrabajo(trabajador, actividad))
            
def reinsercion(inicio,env, trabajador):
    persona= random.randint(0,5)
    ##Determina el area del trabajador
    global actividad
    if persona == 0:
        actividad = "alimentacion"
    elif persona == 1:
        actividad = "retail"
    elif persona == 2:
        actividad = "restaurante y ocio"
    elif persona == 3:
        actividad = "viajes"
    elif persona == 4:
        actividad = "Hoteles"
    ##Tiempo de recuperacin de 2 a 7 das en el domicilio, se puso de 1 a 2 porque se iba a demorar mucho con todos los datos
    ## pero se podra poner el rango de 2 a 7, pero la salida de datos se demoraria ms
    tiempo= random.randint(1,2)
    time.sleep(tiempo)
    print("El trabajador recuperado ", trabajador, " del area ", actividad, "cumplio los 7 das de descanso ")
    tasa_desempleo= random.randint(0,100)
    if(tasa_desempleo>35):
        yield inicio.env.process(inicio.teletrabajo(trabajador, actividad))
    else:
        print("El trabajador recuperado ", trabajador, " del area ", actividad, "queda desempleado ")
        trabajadores_desempleados[env.now]=trabajadores_desempleados[env.now]+ 1 if env.now in trabajadores_desempleados else  1
    
def ejecutar(env, intervalo_llegada):
    inicio=ReinsercionLaboral(env, intervalo_llegada)
    for i in range(8000):
        yield env.timeout(1)
        env.process(reinsercion(inicio,env,i))
        
env=simpy.Environment()

env.process(ejecutar( env,m))

env.run(until =DIAS_SIMULACION)
print("TERMINO EL PROCESO")
```
